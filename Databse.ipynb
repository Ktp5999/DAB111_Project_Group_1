```python
# Bar plot of job details
plt.figure(figsize=(10, 6))
colors = ('blue','red','pink','green','purple','purple','purple','purple','purple','purple','purple','purple')
(df.job.value_counts(normalize = True)*100).plot.bar(color=colors,title = 'Bar plot for Job')
```

```python
# Plot histogram of Age
plt.figure(figsize=(10, 6))
sns.histplot(df['age'],  bins=10, kde = True)
plt.title('Age Distribution of Clients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
